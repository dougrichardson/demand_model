# Format demand data to hourly

```python
import pandas as pd
```

```python
data_path = "/g/data/w42/dr6273/work/projects/Aus_energy/data/energy_demand/"
```

```python
dem = pd.read_csv(
    data_path+"2010_2020demand.zip",
    header=3,
    parse_dates={'timestamp': ['Date','Time']},
    index_col="timestamp",
    usecols=["Date", "Time", "NSW1", "QLD1", "SA1", "TAS1", "VIC1"]
)
```

```python
dem.head()
```

Group and sum by day

```python
dem_daily = dem.resample("D").sum()
```

Remove last row as is first 30 mins of 2021

```python
dem_daily = dem_daily.iloc[:-1]
```

Rename columns

```python
dem_daily.columns = ["NSW", "QLD", "SA", "TAS", "VIC"]
```

```python
dem_daily
```

Write

```python
dem_daily.to_csv(
    data_path+"daily_demand_2010-2020.csv"
)
```

Also calculate monthly

```python
dem_monthly = dem_daily.resample("M").sum()
```

```python
dem_monthly
```

```python
dem_monthly.to_csv(
    data_path+"monthly_demand_2010-2020.csv"
)
```
